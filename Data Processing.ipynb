```python
df = pd.read_csv("kaggle_survey_2022_responses.csv",header=1)
df.head(5)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df['ID'] = np.arange(df.shape[0])
cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
df.head(5)
```

```python
name_df = {'Study platforms':list(range(6,19)),'First platforms':list(range(19,26)),'ML on research':list(range(27,30)), 'Program language':list(range(31,46)),'IDE':list(range(46,60)), 'Host notebook':list(range(60,76)),'Visual libraries':list(range(76,91)),'ML frameworks':list(range(92,107)), 'ML algorithms':list(range(107,121)), 'CV method':list(range(121,129)),'NLP methods':list(range(129,135)),'Pre-train model':list(range(135,145)),'Important task':list(range(151,159)), 'Cloud Computing':list(range(161,173)), 'Cloud Computing Product':list(range(174,179)), 'Data Storage Product':list(range(179,187)),'Data Product':list(range(187,203)),'BI Tools':list(range(203,218)),'ML Products':list(range(218,231)),'Automated ML tools':list(range(231,239)),'deploy ML platform':list(range(239,251)),'Monitor ML tool':list(range(251,266)), 'ethical AI':list(range(266,275)),'Hardware ML':list(range(275,284)),'Media sources':list(range(285,297))}
```

```python
for col in name_df.keys():
    col_list = name_df[col]
    a = []
    for stt in col_list:
        col_name = cols[stt]
        sub_df = df.loc[df[col_name].notnull(),col_name]
        sub_df = sub_df.reset_index()
        sub_df = sub_df.rename(columns = {col_name: col,'index':'ID'})
        a.append(sub_df)
    b = pd.concat(a, ignore_index=True)
    b.to_csv(col + '.csv', index=False)
```

```python
ta = []
for each in name_df.values():
    ta.extend(each)
tx = list(range(len(cols)))
keep_col = [item for item in tx if item not in ta]
cols_tx = [cols[item] for item in keep_col]
main_df = df[cols_tx]
main_df.to_csv('main_df.csv',index=False)
```
